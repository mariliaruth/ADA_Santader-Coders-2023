# Programação Orientada a Objetos - Aula 3

Implementar um sistema de controle de conta corrente bancária.

A conta bancária possui, no mínimo, os seguintes atributos:

1) Agência (str)

2) Conta (str)

3) Saldo (float)

4) Nome do titular (str)

5) Data de nascimento do titular (date)

6) CPF do titular (str)

Todas as contas correntes iniciam sem saldo (R$ 0,00). Os atributos não devem ser acessados diretamente (observar a melhor forma de garantir o acesso quando necessário). O nome do titular pode ser alterado após o cadastro do titular.

Nessa conta corrente deve ser possível realizar:

1) Depósitos

2) Pagamentos (basta informar o valor a ser pago e debitar da conta)

3) Transferências (de uma conta para outra conta)

4) Saques 

5) Extrato (visualiza as operações realizadas na conta, idependente de dia, mês, etc)

No extrato bancário deve constar:

1) Número da Agência

2) Número da conta

3) Nome do titular

4) CPF do titular

5) Saldo final

6) Movimentações ('E' para entrada, 'S' para saída)

Criar um menu no console para navegar entre as opções desejadas.

`````python
from datetime import datetime

class Titular:
    def __init__(self, nome, nascimento, cpf):
        self._nome = nome
        self._nascimento = nascimento
        self._cpf = cpf

    @property
    def nome(self):
        return self._nome

    @property
    def nascimento(self):
        return self._nascimento

    @property
    def cpf(self):
        return self._cpf

class ContaCorrente:
    def __init__(self, agencia, conta, titular):
        self._agencia = agencia
        self._conta = conta
        self._titular = titular
        self._saldo = 0.0
        self._movimentacoes = []

    @property
    def agencia(self):
        return self._agencia

    @property
    def conta(self):
        return self._conta

    @property
    def titular(self):
        return self._titular

    @property
    def saldo(self):
        return self._saldo

    def depositar(self, valor):
        self._saldo += valor
        self._movimentacoes.append((valor, 'E'))

    def pagar(self, valor):
        if self._saldo >= valor:
            self._saldo -= valor
            self._movimentacoes.append((valor, 'S'))
        else:
            print("Saldo insuficiente para efetuar o pagamento.")

    def transferir(self, valor, outra_conta):
        if self._saldo >= valor:
            self._saldo -= valor
            self._movimentacoes.append((valor, 'S'))
            outra_conta.depositar(valor)
        else:
            print("Saldo insuficiente para efetuar a transferência.")

    def sacar(self, valor):
        if self._saldo >= valor:
            self._saldo -= valor
            self._movimentacoes.append((valor, 'S'))
        else:
            print("Saldo insuficiente para efetuar o saque.")

    def extrato(self):
        print("Extrato Bancário")
        print(f"Agência: {self._agencia}")
        print(f"Conta: {self._conta}")
        print(f"Titular: {self._titular.nome}")
        print(f"CPF: {self._titular.cpf}")
        print(f"Saldo Final: R$ {self._saldo:.2f}")
        print("Movimentações:")
        for valor, tipo in self._movimentacoes:
            print(f"Tipo: {tipo}\nValor: {valor}")
           
`````
